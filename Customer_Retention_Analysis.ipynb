```python
df['1Gender of respondent'].value_counts()
```

```python
sns.countplot(x=df['1Gender of respondent'], data=df)
```

Lets check the Age Group of the custemers with respect to the gender

```python
sns.countplot(x=df['2 How old are you?'], hue=df['1Gender of respondent'])
plt.xticks(rotation=90)
plt.show()
```

21-30 years age group are more online shoppers followed by the 31-40 years and 41-50years. 

Lets look into the city where the online shoppers are more. 

```python
sns.countplot(x=df['3 Which city do you shop online from?'])
plt.xticks(rotation=90)
plt.show()
```

Delhi stands highest followed by Greater Noida, Noida and Bangalore. 

```python
location=pd.crosstab(df['4 What is the Pin Code of where you shop online from?'], df['3 Which city do you shop online from?'], margins=True)
location
```

```python
sns.barplot(data=location)
plt.xticks(rotation=90)
plt.show()
```

Location table shows us the city combined with the pincodes. P1incodewise also Delhi stands higher.

```python
df['5 Since How Long You are Shopping Online ?'].value_counts()
```

```python
shoppingyears=pd.crosstab(df['2 How old are you?'], df['5 Since How Long You are Shopping Online ?'])
shoppingyears
```

Also the people from age group 21 to 50 years are more interested in online shopping compared to above 50 years and less than 20 years. Thiscan be due to financial dependancy.
